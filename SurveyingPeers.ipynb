```python
data = [[17,"Republican","Male","Central/South American Food","Other/Maybe/Kind of/etc..."],
        [17,"Other","Female","Asian Food","Yes"],
        [14,"Republican","Male","Central/South American Food","Yes"],
        [17,"Other","Male","Asian Food","Other/Maybe/Kind of/etc..."],
        [17,"Other","Male","Asian Food","Yes"],
        [18,"Other","Male","African Food","Other/Maybe/Kind of/etc..."],
        [16,"Democratic","Male","African Food","Yes"],
        [16,"Democratic","Male","Asian Food","Other/Maybe/Kind of/etc..."],
        [17,"Republican","Male","Asian Food","Yes"],
        [16,"Other","Other","Asian Food","Other/Maybe/Kind of/etc..."],
        [15,"Other","Male","Other","Other/Maybe/Kind of/etc..."],
        [19,"Democratic","Female","Central/South American Food","Yes"],
        [17,"Republican","Male","Other","Other/Maybe/Kind of/etc..."],
        [17,"Democratic","Female","Asian Food","Yes"],
        [16,"Other","Male","Central/South American Food","Yes"]]
key = {"Age": [14,15,16,17,18,19], 
       "Political Affiliation":["Republican","Democratic","Other"],
       "Gender":["Male","Female","Other/Maybe/Kind of/etc..."],
       "Food":["African Food","Asian Food","Central/South American Food","European/American Food","Other"],
       "Enjoyment":["Yes","No","Other/Maybe/Kind of/etc..."]}
keykey = ["Age","Political Affiliation","Gender","Food","Enjoyment"]
i = 0

#Histograms

for z in keykey:
    print(z.upper(),":")
    for y in range(len(key[z])):
        print("\/",key[z][y],"\/")
        for x in data:
            if x[i] == key[z][y]:
                print("*", end="")
        print()
    print()
    i=i+1
```
