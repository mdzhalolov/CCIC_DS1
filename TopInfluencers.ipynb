```python
import pandas as pd
import seaborn as sb
pd.set_option('float_format', '{:f}'.format)
df=pd.read_csv("Top_Influencers.csv")
df.drop(df.columns[[0,1,6,7,9]],axis="columns",inplace=True)
for x in df:
    df[x]=df[x].replace({'k':'*1e3','m':'*1e6','b':'*1e9'}, regex=True).map(pd.eval).astype("int64")
df.sort_values(by="Total Likes", inplace=True,ascending=False)
print(df)
print(df.describe(include="all"))
```

```python
#Number of Followers vs Influence Score
sb.scatterplot(x="Followers", y="Influence Score", data=df)
sb.regplot(x="Followers", y="Influence Score", data=df)
print(df["Followers"].corr(df["Influence Score"]))
```

```python
#Average Likes vs Influence Score
sb.scatterplot(x="Avg. Likes", y="Influence Score", data=df)
sb.regplot(x="Avg. Likes", y="Influence Score", data=df)
print(df["Avg. Likes"].corr(df["Influence Score"]))
```

```python
#Total Likes vs Influence Score
sb.scatterplot(x="Total Likes", y="Influence Score", data=df)
sb.regplot(x="Total Likes", y="Influence Score", data=df)
print(df["Total Likes"].corr(df["Influence Score"]))
```

```python
#Number of Posts vs Influence Score
sb.scatterplot(x="Posts", y="Influence Score", data=df)
sb.regplot(x="Posts", y="Influence Score", data=df)
print(df["Posts"].corr(df["Influence Score"]))
```

```python
#PD/SB Practice below with cereal data set.
import pandas as pd
import seaborn as sb
df=pd.read_csv("cereal.csv")
df.drop(df.columns[[1,2,3,4,5,6,7,8,10,11,12,13,14]],axis="columns",inplace=True)
df=df[(df[["sugars"]] >= 0).all(1)]
df.sort_values(by="rating", inplace=True)
print(df['sugars'].corr(df['rating']))
print(df.to_string())

sb.scatterplot(x="sugars", y="rating", data=df)
sb.regplot(x="sugars", y="rating", data=df)
```
