```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
#pd.set_option('float_format', '{:f}'.format)
df=pd.read_csv("GDPgrowth.csv")
df.drop(df.columns[[2,3,4,65]],axis="columns",inplace=True)
#df.drop([116,80,42,89,48,47,70,63,73,112,31,65,28,29,62,27,103,91,107,106,49,108,97,96,86,0,51,45,50,1], inplace=True)
df['mean'] = df.mean(axis=1)
df.sort_values(by="1968", inplace=True,ascending=False)
listy = {"Country Name": np.nan, "Country Code":np.nan}
for x in range(1961,2021):
    listy[str(x)] = df.loc[df[str(x)].idxmax()]["Country Name"]
df = df.append(listy, ignore_index=True)
x = df.iloc[266].value_counts().to_dict()
x = dict(list(x.items())[:12])
print(x)

```

```python
yAxis = list(x.keys())
xAxis = list(x.values())
#plt.xticks(rotation=45) # used to rotate labels if needed.
plt.xlabel("How many times they had the highest GDP growth in the world in a year")
plt.ylabel("Countries")
plt.title("Countries with the most occurences in GDP growth per year from 1961-2020")
sb.barplot(x=xAxis,y=yAxis)
```
